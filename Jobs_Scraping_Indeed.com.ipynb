```python
# import libraries
import requests
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

```python
#getting website url and checking status of response from http request
website = 'https://pk.indeed.com/jobs?q=data%20analyst&l=Islamabad&from=searchOnHP&vjk=3f7584cbca27dfb3'
response = requests.get(website)
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# getting the main div where all list items are stored
results = soup.find_all('div', {'class' : 'cardOutline'})
len(results)
```

```python
#finding job title
results[0].find('span', {'id':'jobTitle-3f7584cbca27dfb3'}).get_text().strip()
```

```python
#finding company name
results[0].find('span', {'class':'companyName'}).get_text().strip()
```

```python
#finding company location
results[0].find('div', {'class':'companyLocation'}).get_text().strip()
```

```python
results[0].find('div', class_=['heading6 tapItem-gutter metadataContainer noJEMChips salaryOnly']).get_text().strip()
```

```python
results[0].find('div', class_=['job-snippet']).get_text().strip()
```

```python
job_title = []
company_name = []
company_loc = []
salary = []
job_decription = []

for result in results:
    
    
    # job_title
    try:
        job_title.append(result.find('h2', {'class':'jobTitle'}).get_text())
    except:
        job_title.append('n/a')
    
    # company_name
    try:
        company_name.append(result.find('span', {'class':'companyName'}).get_text().strip())
    except:
        company_name.append('n/a')
        
    # company_loc
    try:
        company_loc.append(result.find('div', {'class':'companyLocation'}).get_text())
    except:
        company_loc.append('n/a')
    
    # salary
    try:
        salary.append(result.find('div', {'class':'heading6 tapItem-gutter metadataContainer noJEMChips salaryOnly'}).get_text())
    except:
        salary.append('n/a')
    
    #job_decription 
    try:
        job_decription.append(result.find('div', {'class':'job-snippet'}).get_text().strip())
    except:
        job_decription.append('n/a')
```

```python
# dictionary
jobs = pd.DataFrame({'company_name': company_name, 'job_title':job_title, 'company_location':company_loc,
                                'salary': salary, 'job_decription': job_decription})
```

```python
jobs
```

```python
#data cleaning
jobs['salary'] = jobs['salary'].apply(lambda x: x.strip('+').strip('1'))


jobs['pay'] = jobs['salary'].str.split(' ', expand=True)[1]
jobs['part/full_time'] = jobs['salary'].str.split(' ', expand=True)[0]


jobs['part/full_time'] = jobs['part/full_time'].str.replace('Rs','n/a')
```

```python
jobs
```

```python
jobs.to_excel('indeed_jobs.xlsx', index=False)
```
